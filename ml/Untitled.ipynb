```python
texts = [['human', 'interface', 'computer'],
             ['survey', 'user', 'computer', 'system', 'response', 'time'],
             ['eps', 'user', 'interface', 'system'],
             ['system', 'human', 'system', 'eps'],
             ['user', 'response', 'time'],
             ['trees'],
             ['graph', 'trees'],
             ['graph', 'minors', 'trees'],
             ['graph', 'minors', 'survey']]
```

```python
from gensim.models.doc2vec import TaggedDocument, Doc2Vec
import numpy as np
import random
```

```python
docs = [TaggedDocument(d, [i]) for i, d in enumerate(texts)]
```

```python
def train(doc ,size = 300, min_count=2, batch_size = 10000, epoch_num=10):
    model = Doc2Vec(size=size, min_count=min_count)
    model.build_vocab(docs)
    model.train(docs, total_examples=model.corpus_count, epochs=epoch_num)

    return model
```

```python
model = train(docs)
```

```python
import pandas as pd
pd.DataFrame(list(model.docvecs))
```

```python
inferred_vector = model.infer_vector(['human', 'interface', 'computer'])
```

```python
sims = model.docvecs.most_similar([inferred_vector], topn=len(model.docvecs))
```

```python
for docid, sim in sims:
    print(docid, sim)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
```

```python

corpus=["I come to China to travel",
    "This is a car polupar in China",
    "I love tea and Apple ",
    "The work is to write some papers in science"]

vectorizer = CountVectorizer()
count = vectorizer.fit_transform(corpus)
wb = vectorizer.get_feature_names()
print(wb)
```

```python
tfidf = TfidfTransformer()
res = tfidf.fit_transform(count)
print(res)
```

```python
type(res)
```

```python
res.indices
```

```python
res.data
```

```python
tag = list(res.indptr)
tag[:-1]
```

```python
dic = {
    'word' : [],
    'doc' : [],
    'tfidf' : [],
}

tag = list(res.indptr)[1:-1]
index = 0
doc_index = 0
for word_index, tfidf in zip(res.indices, res.data):
    if len(tag)>0 and index == tag[0]:
        tag.pop(0)
        doc_index += 1
    dic['word'].append(word_index)
    dic['doc'].append(doc_index)
    dic['tfidf'].append(tfidf)
    index += 1
```

```python
import pandas as pd
df = pd.DataFrame(dic)
df
```

```python
for word, group in df.groupby('word'):
    tmp = group.sort_values(by="tfidf",ascending= False)
    print({
        "word": word,
        "doc_tfidf": [{'doc':doc, 'tfidf':tfidf} for doc, tfidf in zip(tmp['doc'].tolist(), tmp['tfidf'].tolist())]
    })
```
