```python
import pandas as pd
import re
```

```python
# #显示所有列
# pd.set_option('display.max_columns', None)
# #显示所有行
# pd.set_option('display.max_rows', None)
#设置value的显示长度为100，默认为50
pd.set_option('max_colwidth',500)
```

```python
re1 = re.compile(r'</p><p class=Para>')
blank_re = re.compile(r'<P class=Para>|</P>|\u3000')

# text = re.sub(re1, '\n', text)
# text = re.sub(blank_re, '', text).strip()
```

```python
path = 'data/news.txt'
data = list()
with open(path, encoding='gb18030', errors='ignore') as fp:
    for line in fp:
        each = [x if x!='' else None for x in line.split('△')]
        if len(each) == 14:
            data.append(each)
```

```python
df = pd.DataFrame(data, columns=['ID', 'MEDIANAME', 'PUBDATE', 'NEWSTITLE', 'AUTHORS', 'SUBTITLE', 'KEYWORDS', 'ABSTRACT', 'LINKADDRESS', 'REPORTLEVEL', 'STATUS', 'ENTRYTIME', 'UPDATETIME', 'GROUNDTIME'])
df = df.loc[1:]
```

```python
def process_text(text):
    text = str(text)
    text = re.sub(re1, '\n', text)
    text = re.sub(blank_re, '', text).strip()
    return text

df['ABSTRACT'] = df['ABSTRACT'].apply(process_text)
```

```python
df
```

```python
df['PUBDATE'] = pd.to_datetime(df['PUBDATE'],format="%Y/%m/%d %H:%M:%S")
df['ENTRYTIME'] = pd.to_datetime(df['ENTRYTIME'],format="%Y/%m/%d %H:%M:%S")
df['UPDATETIME'] = pd.to_datetime(df['UPDATETIME'],format="%Y/%m/%d %H:%M:%S")
df['GROUNDTIME'] = df['GROUNDTIME'].apply(lambda x : x.strip())
df['GROUNDTIME'] = pd.to_datetime(df['GROUNDTIME'],format="%Y/%m/%d %H:%M:%S")
```
