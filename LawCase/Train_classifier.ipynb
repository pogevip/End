```python
from TextClassifier import train
```

```python
import os

dir = 'data/trainSet/classifier/'

# #一起存储
# data_path = os.path.join(dir, 'train_file_5w')
#按类区分存储
data_path = os.path.join(dir, 'web_train_file_dnn')

class_weight_path = os.path.join(dir, 'web_cls.dic')
```

```python
out_path = dir + 'training/web_TextCNN'
train.train(data_path,
            class_weight_path,
            out_path,
            model_option = 'textcnn')
```

```python
out_path = dir + 'training/web_BiLSTM'
train.train(data_path,
            class_weight_path,
            out_path,
            model_option = 'bilstm')
```

```python
out_path = dir + 'training/web_BiLSTM_att'
train.train(data_path,
            class_weight_path,
            out_path,
            model_option = 'bilstm+att')
```
